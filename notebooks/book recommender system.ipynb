## Popularity Based Recommendation System

##### will display top 100 books with highest avg rating, min number of ratings required to be in top 100 would be 250 ratings

```python
# making a new dataset by combining the ratings and books dataset on the basis of ISBN no. of books

ratings_with_names = ratings.merge(books, on = "ISBN")
```

```python
ratings_with_names.head()
```

```python
# making a df with number of ratings for each book, to filter out books that don't pass our critera of number of ratings.

# we're grouping by title and not isbn number as in this dataset, same book titles have different isbn numbers. reset_index converts the series output into a df
num_ratings_df = ratings_with_names.groupby("Book-Title").count()['Book-Rating'].reset_index()
num_ratings_df.rename(columns = {"Book-Rating" : "numberOfRatings"}, inplace = True)
num_ratings_df
```

```python
# creating a df with average rating of a book title, avg rating = sum of ratings / total number of ratings

avg_ratings_df = ratings_with_names.groupby("Book-Title").mean()['Book-Rating'].reset_index()
avg_ratings_df.rename(columns = {"Book-Rating" : "avg_ratings"}, inplace = True)
avg_ratings_df
```

```python
# merging the num of ratings df and the avg rating df

popularity_df = num_ratings_df.merge(avg_ratings_df, on = "Book-Title")
popularity_df
```

```python
# dropping the books where numberOfRatings < 250

# popularity_df[popularity_df["numberOfRatings"] >= 250]
popularity_df = popularity_df[popularity_df["numberOfRatings"] >= 250].sort_values("avg_ratings", ascending = False).head(100)
```

```python
popularity_df
```

```python
# adding more data to our popularity df by merging it with books df

# as some books with same title have different ISBN numbers, our df will increase in size, so we'll drop the duplicates
popularity_df = popularity_df.merge(books, on = "Book-Title").drop_duplicates('Book-Title')[["Book-Title", "Book-Author", "Image-URL-M", "numberOfRatings", "avg_ratings"]]
```

## Collaborative Filtering Based Recommender System

```python
x = ratings_with_names.groupby('User-ID').count()["Book-Rating"] >= 200
# boolean indexing
regular_readers = x[x].index
regular_readers
```

```python
# filtering in our ratings with names dataframe, fetching users which satisfy the upper condition

filtered_ratings = ratings_with_names[ratings_with_names["User-ID"].isin(regular_readers)]
filtered_ratings
```

```python
y = filtered_ratings.groupby("Book-Title").count()["Book-Rating"] >= 50

famous_books = y[y].index
famous_books
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
final_ratings
```

```python
final_ratings.drop_duplicates()
```

```python
# making a pivot table
pt = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values = "Book-Rating")
pt
```

```python
pt.fillna(0, inplace = True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score[0]
```

```python
def recommend(book_name):
    # fetching index with book name:
    suggestions = []
    index = np.where(pt.index == book_name)[0][0]
    suggested_books = sorted(list(enumerate(similarity_score[index])), key = lambda x : x[1], reverse = True)[1 : 6]
    
    for i in suggested_books:
        items = []
        temp_df = books[books["Book-Title"] == pt.index[i[0]]]
        items.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Title"].values))
        items.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Author"].values))
        items.extend(list(temp_df.drop_duplicates("Book-Title")["Image-URL-M"].values))
        
        suggestions.append(items)
        
    return suggestions
```

```python
recommend("1984")
```
